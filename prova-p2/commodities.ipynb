```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import pytz
import investpy as inv
import yfinance as yf
import plotly.graph_objects as go
```

```python
datetime.now()
```

```python
dia_segunda_inv = (datetime.now(pytz.timezone("America/Sao_Paulo"))-timedelta(days=2)).strftime('%d/%m/%Y')
dia_quarta_inv = (datetime.now(pytz.timezone("America/Sao_Paulo"))).strftime('%d/%m/%Y')
dia_quarta_inv
```

```python
df_comm = inv.commodities.get_commodities(group=None)
df_comm.head(20)
```

```python
inv.commodities.get_commodities_overview('grains', as_json=False, n_results=5)
```

```python
df_comm_usd = df_comm[df_comm.currency=='USD']
ls_comm_nomes = list(df_comm_usd.name)
ls_comm_nomes[0:5]
```

```python
df_comm_todas = pd.DataFrame()

for i in ls_comm_nomes:
  ohlc_comm_ind = inv.commodities.get_commodity_historical_data(i, from_date=dia_segunda_inv, to_date=dia_quarta_inv,
                                              country=None, as_json=False, order='ascending', interval='Daily')
  ohlc_comm_ind['Ticker'] = i
  df_comm_todas = pd.concat([df_comm_todas, ohlc_comm_ind], axis=0)
```

```python
df_comm_todas
```

```python
df_comm_open = pd.pivot_table(df_comm_todas, index='Ticker', values='Open', aggfunc='first')
df_comm_high = pd.pivot_table(df_comm_todas, index='Ticker', values='High', aggfunc='max')
df_comm_low = pd.pivot_table(df_comm_todas, index='Ticker', values='Low', aggfunc='min')
df_comm_close = pd.pivot_table(df_comm_todas, index='Ticker', values='Close', aggfunc='last')

df_comm_results = pd.concat([df_comm_open, df_comm_high, df_comm_low, df_comm_close], axis=1)
df_comm_results['Resultado_%'] = (df_comm_results.Close - df_comm_results.Open)/df_comm_results.Open*100
df_comm_results.head(10)
```

```python
df = df_comm_results.style.format('{:.2f}', subset=['Resultado_%']).applymap(
        lambda x: 'font-weight: bold; color: white; background-color : #d40202' if x<-1 else \
        ('font-weight: bold; color: white; background-color : #d66363' if x<0 else \
        ('font-weight: bold; color: white; background-color : #4eb55c' if x>1 else \
        'font-weight: bold; color: black; background-color : #c1dec5')), subset=["Resultado_%"]).\
        set_table_styles([{'selector': 'td', 'props': 'text-align: center'},
        {'selector': 'th', 'props': 'text-align: center; font-weight: bold; background-color: #b8b4b4; color: black; font-size: 14pt'},
        {'selector': 'th.index_name', 'props': 'text-align: center; font-weight: bold; background-color: blue; color: black; font-size: 14pt'},
        {'selector': 'th.row_heading', 'props': 'text-align: center; font-weight: bold; background-color: #f3f70a; color: black; font-size: 11pt'},
        {'selector': 'td:hover', 'props': 'text-align: center; font-weight: bold; background-color: #f3f70a; color: black; font-size: 15pt'},
        ], overwrite=False)
```

```python
df_fig = df_comm_results.sort_values(by='Resultado_%',ascending=False)
df_fig = pd.concat([df_fig.head(5), df_fig.tail(5)], axis=0)

df_fig['Color'] = np.where(df_fig['Resultado_%'] < 0, 'red', 'green')

fig_indices_results = go.Figure()

fig_indices_results.add_trace(go.Bar(name='Net', x=df_fig['Resultado_%'], y=df_fig.index,
                                marker_color=df_fig['Color'], orientation='h'))

annotations = []
x_pos = []
y_pos = []
x_neg = []
y_neg = []

for i in range(0,len(df_fig)):
  
  if df_fig["Resultado_%"].iloc[i] > 0:
    x_pos.append(df_fig.index[i])
    y_pos.append(np.round(df_fig["Resultado_%"].iloc[i], decimals=2))
  
  else:
    x_neg.append(df_fig.index[i])
    y_neg.append(np.round(df_fig["Resultado_%"].iloc[i], decimals=2))

result_max = df_fig["Resultado_%"].max() + 1
result_min = df_fig["Resultado_%"].min() - 1
    
for yd, xd in zip(y_pos, x_pos):
  annotations.append(dict(xref='x1', yref='y1', y=xd, x=yd + 2, text= str(yd)+'%',
                          font=dict(family='Arial', size=17, color='black'), showarrow=False))

for yd, xd in zip(y_neg, x_neg):
  annotations.append(dict(xref='x1', yref='y1', y=xd, x=yd - 2, text= str(yd)+'%',
                          font=dict(family='Arial', size=17, color='black'), showarrow=False))

fig_indices_results.update_xaxes(visible=True, showticklabels=True, range=[-20, 10])
fig_indices_results.update_layout(barmode='stack',title_text='<b>Resultado semanal commodities',
                template = 'simple_white', margin=dict(l=20, r=20, t=70, b=20), paper_bgcolor="#f7f8fa",
                width=800,height=800, annotations=annotations, font=dict(family="Arial",size=17, color = 'black'))
```
